```python
from pathlib import Path

import pandas as pd

csv_dir = Path("../csv").resolve()
paths = sorted(csv_dir.glob("*.csv"))

if not paths:
    raise SystemExit(f"No CSV files found under: {csv_dir}")

dfs = [pd.read_csv(p) for p in paths]
merged = pd.concat(dfs, ignore_index=True, sort=False)

if "place_id" in merged.columns:
    merged = merged.drop_duplicates(subset=["place_id"], keep="first")
else:
    merged = merged.drop_duplicates(keep="first")

#out_path = csv_dir / "merged.csv"
out_path = 'merged.csv'
merged.to_csv(out_path, index=False)

out_path
```
